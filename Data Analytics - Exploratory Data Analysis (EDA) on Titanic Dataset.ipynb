```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Load the Titanic dataset from your file
data = pd.read_csv("C:/Users/Admin/Documents/Titanic-Dataset.csv")

# Show the first few rows
data.head()
```

```python
# Overview of the data
data.info()

# Summary statistics
data.describe()

# Check missing values
print("Missing Values:\n", data.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cbar=False, cmap="YlGnBu")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(data['Age'], kde=True, color='teal')
plt.title("Age Distribution")
plt.show()

# Fare Distribution
plt.figure(figsize=(8,5))
sns.histplot(data['Fare'], kde=True, color='purple')
plt.title("Fare Distribution")
plt.show()
```

```python
# Select numeric columns only
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival by Gender")
plt.show()

# Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title("Survival by Passenger Class")
plt.show()
```
